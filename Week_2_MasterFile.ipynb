# Double Checking Kellen's matching


```python
# This is merging the "Code data" with the emissions data set
# the "Code data is again the file that contains name 'Keys'""
merge_1 = pd.merge(df, code_df, left_on='Organization', right_on = "Name")
df_revenue_merged = pd.merge(merge_1, df_revenue, left_on='source_org_id', right_on = "Company Name")
df_revenue_merged = df_revenue_merged.drop(columns = ['Organization_ID', 'source_org_id', 'Name'])
df_revenue_merged.drop(df_revenue_merged.columns[[0]], axis = 1)
```

#### To me is looks like that matching works well, lets make a new column to compate the amount of emissions to the amount of revenue


```python
# Creating the new column to hold the ratio
df_master = df_revenue_merged
df_master['revenue_vs_emissions'] = (df_master['Revenue (in $USD)'] / df_master['total_scope_3_emissions'])
df_master.head(3)
```

```python
# Sorting the data frame based on the ratio column
worst = df_master.sort_values('revenue_vs_emissions')
worst.head(10)
```
